```python
f1_t1=pd.read_excel('Data 2016.XLSX',sheet_name='Table 1',skiprows=1)
f2_t1=pd.read_excel('Data 2017.XLSX',sheet_name='Table 1',skiprows=2)
f3_t1=pd.read_excel('Data 2018.XLSX',sheet_name='Table 1',skiprows=2)
f1_t1.drop('Unnamed: 7',inplace=True,axis=1)
f2_t1.drop('Unnamed: 7',inplace=True,axis=1)
f3_t1.drop('Unnamed: 7',inplace=True,axis=1)
```

```python
def sp(file):
    ind=[]
    states=[]
    for i in range(len(file.iloc[:,0])):
        if type(file.iloc[:,0][i]) == str:
            ind.append(i)
            states.append(file.iloc[:,0][i])
        
    ind.append(len(file))  
    data={}
    for i in range(0,len(ind)-1):
        s=sum(file.iloc[:,6][ind[i]:ind[i+1]])
        avg=s/(ind[i+1]-ind[i])
        data[states[i]] = round(float(avg),2)
    return data
    
```

```python
so2_2016=sp(f1_t1)
so2_2017=sp(f2_t1)
so2_2018=sp(f3_t1)
```
